```python
import shutil
import os
from os import listdir
from os.path import isfile, join
import pandas as pd
```

```python
image_path = 'C:\BioNerds'
meta_path = 'C:\BioNerds\Metadata_utils'
```

```python
files = [f.strip('.jpg') for f in listdir(image_path) if isfile(join(image_path, f))]
```

```python
meta_csv = pd.read_csv(meta_path + '\HAM10000_metadata.csv')
```

```python
meta_csv.head()
```

```python
for f in files:
    diagnosis = meta_csv[meta_csv['image_id'] == f]['dx'].to_string().strip('1234567890 ')
    if diagnosis == 'nv' :
        shutil.move(image_path + '\\' + f + '.jpg', image_path + r'\1' + '\\')
    if diagnosis == 'bkl' :
        shutil.move(image_path + '\\' + f + '.jpg', image_path + r'\0' + '\\')
    if diagnosis == 'akiec' :
        shutil.move(image_path + '\\' + f + '.jpg', image_path + r'\2' + '\\')
    if diagnosis == 'mel' :
        shutil.move(image_path + '\\' + f + '.jpg', image_path + r'\3' + '\\')
    if diagnosis == 'vasc' :
        shutil.move(image_path + '\\' + f + '.jpg', image_path + r'\4' + '\\')
    if diagnosis == 'bcc' :
        shutil.move(image_path + '\\' + f + '.jpg', image_path + r'\5' + '\\')
    if diagnosis == 'df' :
        shutil.move(image_path + '\\' + f + '.jpg', image_path + r'\6' + '\\')
        
        
    
```
